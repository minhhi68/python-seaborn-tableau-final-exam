```python
# Bỏ 2 cột không có dữ liệu
df_cleaned = df.drop(columns=['Churned', 'Stayed'])
```

```python
# Lọc khách hàng chỉ thuộc "Stayed" hoặc "Churned"
df_active = df_cleaned[df_cleaned['Customer Status'].isin(['Stayed', 'Churned'])]
```

```python
# Thêm cột nhị phân "Churned_Flag"
df_active['Churned_Flag'] = df_active['Customer Status'].apply(lambda x: 1 if x == 'Churned' else 0)
```

```python
# Phân tích tỷ lệ churn theo giới tính
churn_by_gender = df_active.groupby('Gender')['Churned_Flag'].mean().reset_index()
```

```python
# Phân tích churn theo hợp đồng
churn_by_contract = df_active.groupby('Contract')['Churned_Flag'].mean().reset_index()
```

```python
# Biểu đồ 1: Tỷ lệ churn theo giới tính
plt.figure(figsize=(6,4))
sns.barplot(data=churn_by_gender, x='Gender', y='Churned_Flag')
plt.title('Churn Rate by Gender')
plt.ylabel('Churn Rate')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
```

<br>Nhận xét từ biểu đồ </br>

Tỷ lệ churn giữa nam và nữ gần như tương đương:

- Cột biểu diễn cho Male và Female có chiều cao gần bằng nhau.

- Giới tính không phải là yếu tố ảnh hưởng rõ rệt đến khả năng rời bỏ dịch vụ.

- Không có sự khác biệt đáng kể về tỷ lệ churn giữa hai giới, nên việc thiết kế chiến lược giữ chân riêng biệt cho nam hoặc nữ có thể không mang lại hiệu quả vượt trội.

```python
# Biểu đồ 2: Tỷ lệ churn theo loại hợp đồng
plt.figure(figsize=(8,5))
sns.barplot(data=churn_by_contract, x='Contract', y='Churned_Flag')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.xlabel('Contract Type')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

📊 Nhận xét từ biểu đồ
Churn cao nhất ở hợp đồng Month-to-Month:

Tỷ lệ rời đi trong nhóm này trên 50%, cao vượt trội so với các nhóm còn lại.

Đây là nhóm không cam kết dài hạn, nên dễ thay đổi nhà cung cấp dịch vụ khi không hài lòng.

Churn thấp rõ rệt ở hợp đồng One Year và Two Year:

Hợp đồng Two Year thấp nhất trong ba nhóm.

“Dissatisfaction” và “Attitude of Support Personnel” đều chiếm ~17%

Một tỷ lệ đáng kể khách hàng rời đi vì:

Không hài lòng chung với dịch vụ (Dissatisfaction).

Thái độ hỗ trợ kém từ nhân viên.
→ Chất lượng dịch vụ & chăm sóc khách hàng là yếu tố then chốt.

Các lý do còn lại như “Product Dissatisfaction”, “Attitude”, “Price”... chiếm tỷ trọng nhỏ hơn (~5–7%)

Tuy không lớn nhưng vẫn đáng lưu tâm – đặc biệt nếu xảy ra thường xuyên ở một phân khúc cụ thể.



<br>💡 Gợi ý chiến lược kinh doanh: </br>
Phân tích sâu nhóm “Competitor Offer”:

Theo dõi động thái của đối thủ (giá, ưu đãi, dịch vụ).

Tạo chính sách giữ chân linh hoạt như:

Ưu đãi tái ký hợp đồng.

Tặng dịch vụ thêm nếu khách có ý định rời đi.

Cải thiện dịch vụ chăm sóc khách hàng:

Đào tạo lại kỹ năng giao tiếp, thái độ của nhân viên hỗ trợ.

Thiết kế “Churn Prevention Program” theo từng nhóm lý do:

Tự động phát hiện khách hàng có dấu hiệu giống các nhóm lý do này → can thiệp sớm.



```python
# So sánh churn rate theo 2 chiều: loại hợp đồng và dịch vụ Online Security
comparison_df = df_active[df_active['Online Security'].notna()]
```

```python
# Tính tỷ lệ churn theo từng nhóm hợp đồng và bảo mật
grouped = comparison_df.groupby(['Contract', 'Online Security'])['Churned_Flag'].mean().reset_index()
```

```python
# Biểu đồ so sánh có tô màu khác nhau theo Online Security
plt.figure(figsize=(10,6))
sns.barplot(data=grouped, x='Contract', y='Churned_Flag', hue='Online Security')
plt.title('Churn Rate by Contract Type and Online Security')
plt.ylabel('Churn Rate')
plt.xlabel('Contract Type')
plt.legend(title='Online Security')
plt.tight_layout()
plt.show()
```

<br>📊 Nhận xét từ biểu đồ: </br>

Ở mọi loại hợp đồng, khách hàng không dùng Online Security luôn có tỷ lệ churn cao hơn rõ rệt:

Đặc biệt ở nhóm Month-to-Month, chênh lệch rất lớn:

Không có Online Security: churn > 60%

Có Online Security: churn chỉ khoảng ~35–40%

Hiệu quả của Online Security giảm dần theo độ dài hợp đồng:

Với hợp đồng One Year và Two Year, cả hai nhóm (dùng và không dùng Online Security) đều có tỷ lệ churn rất thấp.

→ Vì hợp đồng dài hạn vốn đã giúp giữ chân tốt, nên sự khác biệt do Online Security không còn quá rõ.

Tác động của Online Security mạnh nhất ở nhóm hợp đồng ngắn hạn:

Đây là nhóm khách hàng dễ rời đi → việc có thêm dịch vụ bảo mật giúp tăng sự gắn bó rõ rệt.
