```python
# Bỏ 2 cột không có dữ liệu
df_cleaned = df.drop(columns=['Churned', 'Stayed'])
```

```python
# Lọc khách hàng chỉ thuộc "Stayed" hoặc "Churned"
df_active = df_cleaned[df_cleaned['Customer Status'].isin(['Stayed', 'Churned'])]
```

```python
# Thêm cột nhị phân "Churned_Flag"
df_active['Churned_Flag'] = df_active['Customer Status'].apply(lambda x: 1 if x == 'Churned' else 0)
```

```python
# Phân tích tỷ lệ churn theo giới tính
churn_by_gender = df_active.groupby('Gender')['Churned_Flag'].mean().reset_index()
```

```python
# Phân tích churn theo hợp đồng
churn_by_contract = df_active.groupby('Contract')['Churned_Flag'].mean().reset_index()
```

```python
# Biểu đồ 1: Tỷ lệ churn theo giới tính
plt.figure(figsize=(6,4))
sns.barplot(data=churn_by_gender, x='Gender', y='Churned_Flag')
plt.title('Churn Rate by Gender')
plt.ylabel('Churn Rate')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
```

```python
# Biểu đồ 2: Tỷ lệ churn theo loại hợp đồng
plt.figure(figsize=(8,5))
sns.barplot(data=churn_by_contract, x='Contract', y='Churned_Flag')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.xlabel('Contract Type')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
# Biểu đồ 3: Phân bố churn theo độ tuổi
plt.figure(figsize=(8,5))
sns.histplot(data=df_active, x='Age', hue='Customer Status', multiple='stack', bins=30)
plt.title('Churn Distribution by Age')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

 Insights chính
Hợp đồng ảnh hưởng lớn đến churn:

Khách hàng dùng Month-to-Month có tỷ lệ rời đi cao nhất (51.7%).

Trong khi đó, khách hàng ký Two Year contract có tỷ lệ rời đi rất thấp (2.6%).
→ Gợi ý chiến lược: Khuyến khích chuyển đổi sang hợp đồng dài hạn để giữ chân khách hàng.

Giới tính không phải yếu tố quyết định mạnh:

Tỷ lệ churn giữa nam và nữ gần tương đương, cho thấy giới tính không phải yếu tố gây ảnh hưởng rõ ràng đến hành vi hủy dịch vụ.

Churn phổ biến ở nhóm khách hàng trẻ tuổi hơn:

Biểu đồ phân bố cho thấy khách hàng rời đi tập trung nhiều ở độ tuổi dưới 40.
→ Gợi ý chiến lược: Thiết kế các chương trình giữ chân dành riêng cho nhóm khách trẻ, có thể thích các gói linh hoạt, dịch vụ giải trí.

```python
# 1. Churn rate by tenure group
df_active['Tenure Group'] = pd.cut(df_active['Tenure in Months'],
                                   bins=[0, 12, 24, 36, 48, 60, 72],
                                   labels=['<1yr', '1-2yr', '2-3yr', '3-4yr', '4-5yr', '5-6yr'])
```

```python
# 2. Churn rate by payment method
churn_by_payment = df_active.groupby('Payment Method')['Churned_Flag'].mean().reset_index()
```

```python
# 3. Boxplot: Monthly Charges vs Customer Status
plt.figure(figsize=(8,5))
sns.boxplot(data=df_active, x='Customer Status', y='Monthly Charge')
plt.title('Monthly Charges by Customer Status')
plt.tight_layout()
plt.show()
```

```python
# 4. Barplot: Churn rate by tenure group
plt.figure(figsize=(8,5))
sns.barplot(data=df_active.groupby('Tenure Group')['Churned_Flag'].mean().reset_index(),
            x='Tenure Group', y='Churned_Flag')
plt.title('Churn Rate by Tenure Group')
plt.ylabel('Churn Rate')
plt.xlabel('Tenure Group')
plt.tight_layout()
plt.show()
```

```python
# 5. Barplot: Churn by Payment Method
plt.figure(figsize=(10,6))
sns.barplot(data=churn_by_payment, x='Payment Method', y='Churned_Flag')
plt.title('Churn Rate by Payment Method')
plt.ylabel('Churn Rate')
plt.xlabel('Payment Method')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
# 6. Heatmap: correlation among numerical features
numerical_cols = ['Age', 'Tenure in Months', 'Monthly Charge', 'Total Charges', 
                  'Total Refunds', 'Total Extra Data Charges', 'Total Long Distance Charges', 'Total Revenue']
correlation_matrix = df_active[numerical_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap Between Numerical Features')
plt.tight_layout()
plt.show()
```

Các khía cạnh mới được phân tích
💸 Chi phí hàng tháng và churn:

Khách hàng rời đi (Churned) có xu hướng trả nhiều tiền hơn mỗi tháng so với người ở lại.
→ Gợi ý: Cần rà soát lại mức giá của các gói cao cấp hoặc có thể do các dịch vụ không tương xứng với giá.

📆 Độ dài sử dụng dịch vụ ảnh hưởng lớn:

Khách hàng sử dụng <1 năm có tỷ lệ rời đi gần 60%.

Tỷ lệ này giảm rõ rệt với những người dùng lâu năm.
→ Gợi ý: Tập trung giữ chân khách trong 12 tháng đầu bằng khuyến mãi/thưởng điểm/dịch vụ chăm sóc riêng.

💳 Phương thức thanh toán ảnh hưởng đến churn:

Người dùng Mailed Check (gửi séc) có tỷ lệ churn cao nhất (~41%).

Người dùng Credit Card có churn thấp nhất (~16%).
→ Gợi ý: Khuyến khích chuyển đổi sang thanh toán tự động qua thẻ hoặc ngân hàng.

 Tương quan giữa các chỉ số:

Tenure in Months có tương quan rất mạnh với Total Revenue và Total Charges.

Monthly Charge không hẳn tương quan với Tenure, cho thấy khách mới cũng có thể trả mức phí cao.

```python
# 1. Churn rate theo loại Internet
internet_churn = df_active[df_active['Internet Type'].notna()]
internet_churn_rate = internet_churn.groupby('Internet Type')['Churned_Flag'].mean().reset_index()
```

```python

# 2. Churn theo dịch vụ Online Security
online_security_churn = df_active[df_active['Online Security'].notna()]
security_churn_rate = online_security_churn.groupby('Online Security')['Churned_Flag'].mean().reset_index()
```

```python
# 4. Pie chart: phân bố churn theo Churn Category (loại lý do churn)
churn_category = df_active[df_active['Churn Category'].notna()]
churn_category_counts = churn_category['Churn Category'].value_counts()
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(8,5))
sns.barplot(data=internet_churn_rate, x='Internet Type', y='Churned_Flag')
plt.title('Churn Rate by Internet Type')
plt.ylabel('Churn Rate')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=security_churn_rate, x='Online Security', y='Churned_Flag')
plt.title('Churn Rate by Online Security')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=streaming_churn_rate, x='Streaming TV', y='Churned_Flag')
plt.title('Churn Rate by Streaming TV Usage')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,7))
churn_category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Churn Distribution by Category')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

 Loại Internet và churn
Người dùng Fiber Optic có tỷ lệ rời đi cao nhất (~42%), cao hơn hẳn so với DSL (~20%) và Cable (~28%).
→ Gợi ý: Khách hàng Fiber Optic có thể kỳ vọng cao về tốc độ và dịch vụ, cần tập trung đảm bảo chất lượng.

Dịch vụ bảo mật trực tuyến (Online Security)
Khách hàng không dùng Online Security có tỷ lệ churn cao gấp hơn 3 lần so với người có dùng.
→ Gợi ý: Cung cấp bảo mật miễn phí 3 tháng cho khách hàng mới để giữ chân tốt hơn.

 Dịch vụ Streaming TV
Khách hàng có sử dụng Streaming TV có tỷ lệ rời đi thấp hơn (~31%) so với người không sử dụng (~36%).
→ Gợi ý: Các dịch vụ giải trí giúp tăng sự gắn bó. Có thể tặng gói xem TV cho nhóm có nguy cơ churn cao.

Lý do chính khiến khách hàng rời đi
Theo Pie Chart:

45% churn vì chuyển sang đối thủ.

17% do không hài lòng, 17% do thái độ.
→ Gợi ý:

Cần theo dõi hoạt động của đối thủ, nâng cao dịch vụ khách hàng.

Cải thiện thái độ và hỗ trợ kỹ thuật từ nhân viên, đặc biệt với khách hàng mới và gói cao cấp.

```python
# So sánh churn rate theo 2 chiều: loại hợp đồng và dịch vụ Online Security
comparison_df = df_active[df_active['Online Security'].notna()]
```

```python
# Tính tỷ lệ churn theo từng nhóm hợp đồng và bảo mật
grouped = comparison_df.groupby(['Contract', 'Online Security'])['Churned_Flag'].mean().reset_index()
```

```python
# Biểu đồ so sánh có tô màu khác nhau theo Online Security
plt.figure(figsize=(10,6))
sns.barplot(data=grouped, x='Contract', y='Churned_Flag', hue='Online Security')
plt.title('Churn Rate by Contract Type and Online Security')
plt.ylabel('Churn Rate')
plt.xlabel('Contract Type')
plt.legend(title='Online Security')
plt.tight_layout()
plt.show()
```

```python
# So sánh tổng chi tiêu (Total Charges) giữa nam và nữ, phân theo tình trạng khách hàng
plt.figure(figsize=(10,6))
sns.barplot(data=df_active, x='Gender', y='Total Charges', hue='Customer Status', ci=None)
plt.title('Total Charges by Gender and Customer Status')
plt.ylabel('Total Charges')
plt.xlabel('Gender')
plt.legend(title='Customer Status')
plt.tight_layout()
plt.show()
```

Với mọi loại hợp đồng, khách hàng có sử dụng Online Security luôn có tỷ lệ churn thấp hơn.
